### Inferential Statistics: Stroop Task

#### Background Information
In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participant’s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a congruent words condition, and an incongruent words condition. In the congruent words condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE. In the incongruent words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

#### 2. What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

This task will show the null hypothesis designed such that it will be rejected if there is a significant difference in the mean response time of each sample population for the two conditions under investigation. Random variation explains observable differences in this model. 

Represented mathematically as follows:
    
    H0: μC = μI

Where:
    
    H0 = Null hypothesis.
    μC = The population mean response time under the congruent condition.
    μI = The population mean response time under the incongruent condition.

At initial review the mean for each condition appears fairly different, 14.05 seconds for the congruent condition and 22.02 seconds for the incongruent condition. We will seek to explore the nature of the difference to determine whether chance is the contributing factor or that the results arise from a statistically significant difference between the sample populations. We begin thinking the results are due to random variance and will reject this assumption in the null hypothesis to demonstrate a statistically significant difference occurring in the alternate hypothesis.

The alternate hypothesis is designed to be true if the null hypothesis is rejected. This hypothesis will show the population mean for the response time under congruent condition as significantly less than the population mean for the response time under incongruent condition. 

Represented mathematically as follows: 
    
    Ha: μC < μI

Where:
    
    Ha = Alternate hypothesis.
    μC = The population mean response time under the congruent condition.
    μI = The population mean response time under the incongruent condition.

In initial review of the data set, the reason we include the less-than symbol in the alternate hypothesis is because there is a good chance the congruent condition response times could be significantly less than the incongruent condition response times. This is a directional, one-tailed test. If we were unable to make a distinction one way or the other we could use a different alternate hypothesis requiring the use of a two-tailed non-directional test. The alternate hypothesis for this example would be represented as Ha: μC ≠ μI.

A t-test will be performed on this data instead of a z-test because we are comparing sample means instead of the comparison of population mean to sample mean. We also have no additional details regarding our population parameters. A t-test is also suitable for our investigation given the small sample size and the potential for non-normality in the data distribution.

#### 5. Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

With the consideration that the two treatments, congruent and incongruent conditions, are measured for the same sample population a paired-sample t-test will be used here. We will test whether or not the means for the pairs of observation are significantly different. We will use a confidence interval of 90% and our Degrees of Freedom = 23 (df = n-1). The critical t value is 1.714.

We will determine whether or not the null hypothesis can rejected (if there is a statistical significance between the two conditions) by confirming that the calculated test statistic is greater than the critical t value. The t distribution table will help determine the critical t-value: https://s3.amazonaws.com/udacity-hosted-downloads/t-table.jpg

Now let's Calculate the test statistic.

```python
mean_dif = np.mean(stroop_df['Incongruent']) - np.mean(stroop_df['Congruent'])

print(mean_dif)
```

Above is the mean difference between incongruent and congruent conditions. Next we will divide the difference in means by the estimated standard error of the mean, calculated below.

```python
ind_dif = stroop_df['Congruent'] - stroop_df['Incongruent']

print(ind_dif)
```

```python
sd_dif = np.std(ind_dif, ddof=1)
print(sd_dif)
```

```python
se = sd_dif/np.sqrt(24)
print(se)
```

```python
test_statistic = mean_dif/se
print(test_statistic)
```

As shown in the results of these calculations we see that the test statistic (8.02) is greater than our t value (1.714). Taking into account that this figure has a degrees of freedom of 23 and a confidence interval of 90% we can reject the null hypothesis, concluding that there is a statistically significant difference in the completion time for the task under each condition.
